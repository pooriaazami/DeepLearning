```python
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
num_samples = 100000
counter = 0

Xs = []
Ys = []
labels = []

for _ in range(num_samples):
    x = np.random.rand()
    y = np.random.rand()
    label = x ** 2 + y ** 2 <= 1
    
    counter += label
    
    Xs.append(x)
    Ys.append(y)
    labels.append(label)

pi = counter / num_samples * 4
pi
```

```python
fig, ax = plt.subplots(figsize=(10, 10))
sns.scatterplot(x=Xs, y=Ys, hue=labels, s=.7)
plt.show()
```
