```python
import tensorflow as tf
import tensorflow.keras as keras
import tensorflow.keras.layers as layers

import tensorflow_datasets as tfds

import numpy as np
import matplotlib.pyplot as plt
```

```python
split = ['train', 'test[:50%]', 'test[50%:]']

train_data, validation_data, test_data = tfds.load('horses_or_humans', split=split, as_supervised=True)
```

```python
SIZE = 300

def preprocess(img, label):
  return tf.image.resize(img, [SIZE, SIZE]) / 255., label
```

```python
BATCH_SIZE = 32

train_data = train_data.map(preprocess).batch(BATCH_SIZE)
validation_data = validation_data.map(preprocess).batch(BATCH_SIZE)
test_data = test_data.map(preprocess).batch(BATCH_SIZE)
```

```python
#last layer: Sigmoid

model = keras.Sequential([
    layers.Input(shape=(300, 300, 3), name='input_layer'),

    layers.Conv2D(16, 3, padding='same', activation='relu', name='conv1_1'),
    layers.Conv2D(16, 3, padding='same', activation='relu', name='conv1_2'),
    layers.MaxPool2D(2, name='pool_1'),
    # layers.BatchNoemalization(name='bn_1')

    layers.Conv2D(32, 3, padding='same', activation='relu', name='conv2_1'),
    layers.Conv2D(32, 3, padding='same', activation='relu', name='conv2_2'),
    layers.MaxPool2D(2, name='pool_2'),
    # layers.BatchNoemalization(name='bn_2')

    layers.Conv2D(64, 3, padding='same', activation='relu', name='conv3_1'),
    layers.Conv2D(64, 3, padding='same', activation='relu', name='conv3_2'),
    layers.MaxPool2D(2, name='pool_3'),
    # layers.BatchNoemalization(name='bn_3')

    layers.Flatten(name='flatten'),

    layers.Dense(256, activation='relu', name='fc1'),
    layers.Dense(1, activation='sigmoid', name='output_layer')
])
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
history = model.fit(train_data, validation_data=validation_data, epochs=10)
```
